# 01 Import Libraries

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# Add the file directory to sys.path
results_dir = ''
```

# 02 Load NGS results

```python
csv_file_path = results_dir + 'NGS_results_exact.csv'
NGS_results_exact = pd.read_csv(csv_file_path)
num_itr_NGS_exact = np.array(NGS_results_exact['num_itr'])
sup_err_NGS_exact = np.array(NGS_results_exact['sup_err'])
```

# 03 Plot NGS vs LSP fixed bases

```python
csv_file_path = results_dir + 'LSP_results_exact_fixed_basis_legendre_weighted_avg.csv'
LSP_results_exact = pd.read_csv(csv_file_path)
num_itr_LSP = np.array(LSP_results_exact['num_itr'])
sup_err_LSP_5 = np.array(LSP_results_exact['sup_err_5'])
sup_err_LSP_7 = np.array(LSP_results_exact['sup_err_7'])
sup_err_LSP_9 = np.array(LSP_results_exact['sup_err_9'])
```

```python
# plot result of learning the solution path for every 100 gradient calls
plt.figure(figsize=(18,18))
plt.plot((num_itr_NGS_exact[:len(num_itr_NGS_exact)-4]), np.log(sup_err_NGS_exact[:len(num_itr_NGS_exact)-4]), '-o', linewidth=30, markersize=20, label = "NGS", alpha=0.5)
plt.plot((num_itr_LSP), np.log(sup_err_LSP_5), '--.', linewidth=5, markersize=13, label = "LSP-5 Legendre")
plt.plot((num_itr_LSP), np.log(sup_err_LSP_7), '--.', linewidth=5, markersize=13, label = "LSP-7 Legendre")
plt.plot((num_itr_LSP), np.log(sup_err_LSP_9), '--.', linewidth=5, markersize=13, label = "LSP-9 Legendre")

# Set global font size
mpl.rcParams['font.size'] = 55

font_size = 55
# Tick labels
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.xlim(0, )
# Legend, if applicable
plt.legend(fontsize=font_size)

plt.xlabel('# Passes', fontsize=font_size)
plt.ylabel('Solution Path Error \n Presented in $\\log_{10}$', fontsize=font_size)
plt.grid(True)

# Show the plot
# plt.show()

plt.savefig("itr_exact.png")
```

# 04 Plot Weighted Averaged Iterates vs Last Iterates

```python
csv_file_path = results_dir + 'LSP_results_exact_fixed_basis_legendre_last_iterate.csv'
LSP_last_iterate = pd.read_csv(csv_file_path)
num_itr = np.array(LSP_last_iterate['num_itr'])
sup_err_5_last_iterate = np.array(LSP_last_iterate['sup_err_5'])
sup_err_7_last_iterate = np.array(LSP_last_iterate['sup_err_7'])
sup_err_9_last_iterate = np.array(LSP_last_iterate['sup_err_9'])
```

```python
csv_file_path = results_dir + 'LSP_results_exact_fixed_basis_legendre_last_iterate_shrink_lr.csv'
LSP_last_iterate_shrink_lr = pd.read_csv(csv_file_path)
num_itr = np.array(LSP_last_iterate_shrink_lr['num_itr'])
sup_err_5_last_iterate_shrink_lr = np.array(LSP_last_iterate_shrink_lr['sup_err_5'])
sup_err_7_last_iterate_shrink_lr = np.array(LSP_last_iterate_shrink_lr['sup_err_7'])
sup_err_9_last_iterate_shrink_lr = np.array(LSP_last_iterate_shrink_lr['sup_err_9'])
```

```python
csv_file_path = results_dir + 'LSP_results_exact_fixed_basis_legendre_weighted_avg.csv'
LSP_weighted_avg = pd.read_csv(csv_file_path)
num_itr = np.array(LSP_weighted_avg['num_itr'])
sup_err_5_weighted_avg = np.array(LSP_weighted_avg['sup_err_5'])
sup_err_7_weighted_avg = np.array(LSP_weighted_avg['sup_err_7'])
sup_err_9_weighted_avg = np.array(LSP_weighted_avg['sup_err_9'])
```

```python
# plot result of learning the solution path for every 100 gradient calls
plt.figure(figsize=(18,18))
# plt.plot((num_itr), (sup_err_5_last_iterate), '-', linewidth=13, markersize=15, label = "last iterate w/o shrink", alpha=0.4)
plt.plot((num_itr), (sup_err_7_last_iterate), '-', linewidth=13, markersize=15, label = "LSP-7 last itr w/o shrink", alpha=0.4)
plt.plot((num_itr), (sup_err_9_last_iterate), '-', linewidth=13, markersize=15, label = "LSP-9 last itr w/o shrink", alpha=0.4)

# plt.plot((num_itr), (sup_err_5_last_iterate_shrink_lr), '--.', linewidth=7, markersize=14, label = "last iterate shrink lr", alpha=0.6)
plt.plot((num_itr), (sup_err_7_last_iterate_shrink_lr), '--.', linewidth=7, markersize=14, label = "LSP-7 last itr shrink lr", alpha=0.6)
plt.plot((num_itr), (sup_err_9_last_iterate_shrink_lr), '--.', linewidth=7, markersize=14, label = "LSP-9 last itr shrink lr", alpha=0.6)

# plt.plot((num_itr), (sup_err_5_weighted_avg), '--.', linewidth=7, markersize=14, label = "weighted avg shrink lr")
plt.plot((num_itr), (sup_err_7_weighted_avg), '--.', linewidth=7, markersize=14, label = "LSP-7 wghtd avg shrink lr")
plt.plot((num_itr), (sup_err_9_weighted_avg), '--.', linewidth=7, markersize=14, label = "LSP-9 wghtd avg shrink lr")


# Set global font size
mpl.rcParams['font.size'] = 55

font_size = 55

# Tick labels
plt.yscale('log')

ticks = np.arange(-6.5, 1.5, 1, dtype=float)  # Tick positions in powers of 10
tick_labels = [f'$10^{{{t}}}$' for t in ticks]  # LaTeX-formatted labels

# Set the y-ticks and labels

plt.xticks(fontsize=font_size)
# plt.yticks(fontsize=font_size)
plt.yticks(10**(ticks), tick_labels, fontsize=font_size)
plt.xlim(0, )
# Legend, if applicable
plt.legend(fontsize=font_size)

plt.xlabel('Gradient Calls', fontsize=font_size)
plt.ylabel('Solution Path Error', fontsize=font_size)
plt.grid(True)

# Show the plot
# plt.show()

plt.savefig("last_itr_vs_weighted_avg.png")
```

# 06 Plot fixed and adaptive together

```python
csv_file_path = results_dir + 'LSP_results_exact_fixed_basis_legendre_last_iterate_shrink_lr.csv'
LSP_results_exact = pd.read_csv(csv_file_path)
num_itr_LSP = np.array(LSP_results_exact['num_itr'])
sup_err_LSP_5 = np.array(LSP_results_exact['sup_err_5'])
sup_err_LSP_7 = np.array(LSP_results_exact['sup_err_7'])
sup_err_LSP_9 = np.array(LSP_results_exact['sup_err_9'])
```

```python
csv_file_path = results_dir + 'LSP_results_exact_boosted_legendre.csv'
LSP_results_exact_boosted = pd.read_csv(csv_file_path)
num_itr_LSP_boosted = np.array(LSP_results_exact_boosted['num_itr'])
sup_err_LSP_boosted = np.array(LSP_results_exact_boosted['sup_err'])
```

```python
csv_file_path = results_dir + 'LSP_results_exact_boosted_breaks_legendre.csv'
LSP_results_exact_boosted_breaks = pd.read_csv(csv_file_path)
breaks = np.array(LSP_results_exact_boosted_breaks['breaks'])
```

```python
# plot result of learning the solution path for every 100 gradient calls
plt.figure(figsize=(18,18))
plt.plot((num_itr_NGS_exact[:len(num_itr_NGS_exact)-4]), (sup_err_NGS_exact[:len(num_itr_NGS_exact)-4]), '--o', linewidth=25, markersize=30, label = "Unif Discrtztn", alpha=0.5, zorder=0)
for break_itr in breaks:
    plt.axvline(x = break_itr, color='tab:orange', alpha=0.6)
plt.plot((num_itr_LSP_boosted), (sup_err_LSP_boosted), '--o', linewidth=8, markersize=20, label = "ALSP", alpha=0.8, zorder=2)
plt.plot((num_itr_LSP), (sup_err_LSP_5), '--.', linewidth=5, markersize=15, label = "LSP (q=5)", alpha=0.7, zorder=1)
plt.plot((num_itr_LSP), (sup_err_LSP_7), '--.', linewidth=5, markersize=15, label = "LSP (q=7)", alpha=0.7, zorder=1)
plt.plot((num_itr_LSP), (sup_err_LSP_9), '--.', linewidth=5, markersize=15, label = "LSP (q=9)", alpha=0.7, zorder=1)

# Set global font size
mpl.rcParams['font.size'] = 55

font_size = 55

# Tick labels
plt.yscale('log')

# ticks = np.arange(-6, 1, 1, dtype=float)  # Tick positions in powers of 10
# tick_labels = [f'$10^{{{t}}}$' for t in ticks]  # LaTeX-formatted labels

# Set the y-ticks and labels

plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
# plt.yticks(10**(ticks), tick_labels, fontsize=font_size)
plt.xlim(0, )
# Legend, if applicable
plt.legend(fontsize=font_size, loc=1)

plt.xlabel('Gradient Calls', fontsize=font_size)
plt.ylabel('Solution Path Error', fontsize=font_size)
plt.grid(True)

# Show the plot
# plt.show()

plt.savefig("itr_exact.png")
```
