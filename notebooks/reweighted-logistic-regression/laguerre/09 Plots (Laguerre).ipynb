# 01 Import Libraries

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
# Add the file directory to sys.path
# results_dir = '/content/Learning-the-Optimal-Solution-Path/experiments/fair-regression/results/'
results_dir = ''
```

# 02 Load NGS results

```python
csv_file_path = results_dir + 'NGS_results_exact_laguerre.csv'
NGS_results_exact = pd.read_csv(csv_file_path)
num_itr_NGS_exact = np.array(NGS_results_exact['num_itr'])
sup_err_NGS_exact = np.array(NGS_results_exact['sup_err'])
```

# 03 Plot NGS vs LSP fixed bases

```python
csv_file_path = results_dir + 'LSP_results_exact_fixed_basis_laguerre.csv'
LSP_results_exact = pd.read_csv(csv_file_path)
num_itr_LSP = np.array(LSP_results_exact['num_itr'])
sup_err_LSP_5 = np.array(LSP_results_exact['sup_err_5'])
sup_err_LSP_7 = np.array(LSP_results_exact['sup_err_7'])
sup_err_LSP_9 = np.array(LSP_results_exact['sup_err_9'])
```

```python
# plot result of learning the solution path for every 100 gradient calls
plt.figure(figsize=(20,20))
plt.plot((num_itr_NGS_exact[:len(num_itr_NGS_exact)-3]), np.log10(sup_err_NGS_exact[:len(num_itr_NGS_exact)-3]), '-o', linewidth=30, markersize=20, label = "Naive Grid Search", alpha=0.5)
plt.plot((num_itr_LSP), np.log10(sup_err_LSP_5), '--.', linewidth=5, markersize=13, label = "LSP-5 Laguerre")
plt.plot((num_itr_LSP), np.log10(sup_err_LSP_7), '--.', linewidth=5, markersize=13, label = "LSP-7 Laguerre")
plt.plot((num_itr_LSP), np.log10(sup_err_LSP_9), '--.', linewidth=5, markersize=13, label = "LSP-9 Laguerre")

# Set global font size
mpl.rcParams['font.size'] = 55

font_size = 55
# Tick labels
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.xlim(0, )

# Legend, if applicable
plt.legend(fontsize=font_size)

plt.xlabel('# Passes', fontsize=font_size)
plt.ylabel('Solution Path Error \n Presented in $\\log_{10}$', fontsize=font_size)
plt.grid(True)

# Show the plot
# plt.show()

plt.savefig("itr_exact.png")
```

# 04 Plot NGS vs LSP adaptive basis

```python
csv_file_path = results_dir + 'LSP_results_exact_boosted_laguerre.csv'
LSP_results_exact_boosted = pd.read_csv(csv_file_path)
num_itr_LSP_boosted = np.array(LSP_results_exact_boosted['num_itr'])
sup_err_LSP_boosted = np.array(LSP_results_exact_boosted['sup_err'])
```

```python
csv_file_path = results_dir + 'LSP_results_exact_boosted_breaks_laguerre.csv'
LSP_results_exact_boosted_breaks = pd.read_csv(csv_file_path)
breaks = np.array(LSP_results_exact_boosted_breaks['breaks'])
```

```python
# plot result of learning the solution path for every 100 gradient calls
plt.figure(figsize=(18,18))
plt.plot((num_itr_NGS_exact), np.log10(sup_err_NGS_exact), '-o', linewidth=30, markersize=20, label = "Naive Grid Search", alpha=0.5)
for break_itr in breaks:
    plt.axvline(x = break_itr, color='tab:orange', alpha=0.5)
plt.plot((num_itr_LSP_boosted), np.log10(sup_err_LSP_boosted), '--o', linewidth=8, markersize=20, label = "ALSP Laguerre")

# Set global font size
mpl.rcParams['font.size'] = 55

font_size = 55
# Tick labels
plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
plt.xlim(0, )

# Legend, if applicable
plt.legend(fontsize=font_size)

plt.xlabel('# Passes', fontsize=font_size)
plt.ylabel('Solution Path Error \n Presented in $\\log_{10}$', fontsize=font_size)
plt.grid(True)

# Show the plot
# plt.show()

plt.savefig("boosted_exact_laguerre.png")
```

# 07 Plot of dampened vs standard Laguerre

```python
from scipy.special import eval_laguerre
```

```python
lambdas = np.linspace(0, 20, 200)
```

```python
# plot result of learning the solution path for every 100 gradient calls
plt.figure(figsize=(18,18))

for i in range(1, 6):
    plt.plot((lambdas), (np.sqrt(10) * np.exp(-0.45 * lambdas) * eval_laguerre(i, lambdas)), '-', linewidth=7, label = f"degree {i}")

# Set global font size
mpl.rcParams['font.size'] = 55

font_size = 55

# # Tick labels
# plt.yscale('log')

# ticks = np.arange(-2.5, .5, .5, dtype=float)  # Tick positions in powers of 10
# tick_labels = [f'$10^{{{t}}}$' for t in ticks]  # LaTeX-formatted labels

# Set the y-ticks and labels

plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
# plt.yticks(10**(ticks), tick_labels, fontsize=font_size)
plt.xlim(0, 20)
plt.ylim(-5, 5)
# Legend, if applicable
plt.legend(fontsize=font_size)

plt.ylabel('$L_n(\\lambda)$', fontsize=font_size)
plt.xlabel('$\\lambda$', fontsize=font_size)
plt.grid(True)

# Show the plot
# plt.show()

plt.savefig("dampened_laguerre.png")
```

```python
# plot result of learning the solution path for every 100 gradient calls
plt.figure(figsize=(18,18))

for i in range(1, 6):
    plt.plot((lambdas), (eval_laguerre(i, lambdas)), '-', linewidth=7, label = f"degree {i}")

# Set global font size
mpl.rcParams['font.size'] = 55

font_size = 55

# # Tick labels
# plt.yscale('log')

# ticks = np.arange(-2.5, .5, .5, dtype=float)  # Tick positions in powers of 10
# tick_labels = [f'$10^{{{t}}}$' for t in ticks]  # LaTeX-formatted labels

# Set the y-ticks and labels

plt.xticks(fontsize=font_size)
plt.yticks(fontsize=font_size)
# plt.yticks(10**(ticks), tick_labels, fontsize=font_size)
plt.xlim(0, 20)
plt.ylim(-15, 15)
# Legend, if applicable
plt.legend(fontsize=font_size)

plt.ylabel('$L_n(\\lambda)$', fontsize=font_size)
plt.xlabel('$\\lambda$', fontsize=font_size)
plt.grid(True)

# Show the plot
# plt.show()

plt.savefig("standard_laguerre.png")
```
