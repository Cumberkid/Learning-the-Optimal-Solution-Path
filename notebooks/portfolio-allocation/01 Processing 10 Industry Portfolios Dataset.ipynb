```python
import numpy as np
import pandas as pd
```

```python
portfolios = pd.read_csv('10_Industry_Portfolios_10_Year_Monthly.csv')
portfolios.info()
```

```python
mean = portfolios.mean()
mean.T
```

```python
mean.to_csv('mean.csv', index=False)
```

```python
cov = portfolios.cov()
cov
```

```python
pd.DataFrame(cov).describe()
```

```python
# Compute the eigenvalues of covariance matrix
eigenvalues = np.linalg.eigvals(cov)

# Print the eigenvalues
print("Eigenvalues of covariance matrix:")
print(eigenvalues)
```

```python
I = np.eye(cov.shape[0])

# regularize covariance matrix
pert_cov = cov + 20 * I
pert_cov
```

```python
# Compute the eigenvalues of covariance matrix
eigenvalues = np.linalg.eigvals(pert_cov)

# Print the eigenvalues
print("Eigenvalues of perturbed covariance matrix:")
print(eigenvalues)
```

```python
# Perform Cholesky decomposition to get upper triangular matrix L
decomp_cov = np.linalg.cholesky(cov).T
```

```python
pd.DataFrame(decomp_cov).to_csv('decomp_cov.csv', index=False)
```
